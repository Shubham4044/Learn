```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.__version__
sns.set(style="ticks", color_codes=True)
#style="ticks": This parameter sets the overall style of the plots. In this case, the style is set to "ticks".
#This style is characterized by having ticks along the x-axis and y-axis.
```

```python
#https://seaborn.pydata.org/tutorial.html
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

# Plotting univariate distributions

```python
#Histograms
```

```python
sns.distplot(titanic["survived"]);
#Note: sns.distplot() has been deprecated since Seaborn version 0.11.0. 
#The recommended alternatives are sns.histplot() and sns.kdeplot().
```

```python
sns.distplot(titanic["survived"], kde=False, rug=False);
#A rug plot shows small vertical tick marks along the x-axis, corresponding to the exact data points in the dataset.
#By default, the rug plot is not displayed. If rug=True, small tick marks would appear below the histogram bars to indicate where each data point lies.
```

```python
#Kernel Density Estimation Plots (KDE)
```

```python
sns.distplot(titanic["pclass"], hist=False, rug=True);
```

A kernel density estimate (KDE) is a non-parametric way to estimate the probability density function (PDF) of a continuous random variable. It provides a smooth, continuous representation of the distribution of data, which is useful for visualizing and analyzing the underlying probability distribution. It gives a sense of where the data is concentrated (peaks in the distribution) and how it is spread.
1. Kernel Function: KDE uses a kernel function, typically a Gaussian (normal) kernel, which is a smooth, bell-shaped curve. The kernel function is centered at each data point in the dataset.

2. Smoothing Parameter (Bandwidth): The bandwidth parameter determines the width of the kernel. A larger bandwidth results in a smoother KDE plot, while a smaller bandwidth produces a more detailed, but potentially noisy, plot.

3. Estimation: At each point in the domain (usually the range of the data), the kernel functions centered at each data point are summed up. The result is a continuous estimate of the PDF, where the contribution of each data point to the estimated density decreases as the distance from the point increases.

```python
sns.kdeplot(titanic["pclass"])
sns.kdeplot(titanic["survived"], bw=.2, label="survived")
plt.legend();
```

# Plotting bivariate distributions

```python
#jointplot
```

```python
sns.jointplot(x=titanic["age"], y=titanic["pclass"], data=titanic);
```

```python
#Hexbin plots
```

```python
#Seaborn provides pre-configured styles for plots, like "darkgrid", "whitegrid", "dark", "white", and "ticks".
#"white" removes gridlines and adds a clean, minimalistic look to the plot, making the axes blend into the background.
with sns.axes_style("white"):
    sns.jointplot(x=titanic["age"], y=titanic["pclass"], kind="hex", color="k");
#"k" stands for black.
```

"hex" creates a hexbin plot, which groups points into hexagonal bins and uses color intensity to represent the number of observations in each bin.
Hexbin plots are particularly useful for visualizing dense datasets, where individual points may overlap in a scatter plot.


# Visualizing pairwise relationships in a dataset

```python
mod_titanic = pd.DataFrame()
mod_titanic["survived"] = titanic["survived"]
mod_titanic["pclass"] = titanic["pclass"]
mod_titanic["age"] = titanic["age"]
mod_titanic.head()
```

```python
sns.pairplot(mod_titanic);
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris);
```
